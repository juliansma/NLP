## Escoger filas

Podemos escoger ciertas filas a través de su posición. utilizando *.iloc[posición de la fila que queremos,]*

```python
print(iris.iloc[0:1,])
print(type(iris.iloc[0:1,]))
```

La función .iloc también nos permite especificar filas y columnas separando con la coma el índice de la fila y el de la columna. En el ejemplo siguiente se escogen las filas 8 a la 11 (el intervalo es abierto en el límite superior) y las columnas 1 y 2. Hay que tener en cuenta que los índices en python y también en pandas comienzan en el 0.

```python
print(iris.iloc[8:12,1:3])
```

También escoger filas con base a alguna condicion o característica específica usando un simbolo lógico.

```python
iris[iris["sepal_width"] > 3.5] 
```

Filas que no tengan valores faltantes en determinada columna

```python
no_na = iris[iris["sepal_width"].notna()]
print(no_na.shape)
print(iris.shape)
```

Para este caso no tenemos valores nulos.

### Filas con valores especificos

Existen varias maneras de realizar estas búsquedas, algunas de ellas son utilizando *isin*, *loc* de la librería seaborn y *or*.

```python
codiris = iris[iris["sepal_width"].isin([3.6])]
codiris
```

```python
codiris2 = iris.loc[:, 'petal_width'] == 0.1
codiris2_ = iris.loc[codiris2]
codiris2_
```

```python
iris[(iris['petal_width'] == 0.1) | (iris["petal_width"] == 0.2)]
```

Filas con más de una característica específica.

```python
iris[(iris["sepal_width"] == 3) & (iris["sepal_width"] == 3.7)].info()
```

## Manipulación datos textuales (Limpieza)

Para realizarle limpieza a una columna de nuestra base de datos podemos utilizar los siguientes metodos:

* **lower:**Poner en minúscula todo el texto.
* **replace:**Remplazar ciertos valores por otros.
* **strip:**Eliminar los espacios al principio y al final de la cadena.
* **title:**Poner primera letra de cada palabra en mayúscula.

## **Creación de columnas a partir de otras**

```python
iris["area_sepal"]= iris["sepal_length"]*iris["sepal_width"]
print(iris[["area_sepal","sepal_length","sepal_width"]])
```

### Renombrar columnas

```python
print(iris.columns)
iris.rename(columns={"area_sepal":"sepal_area"},inplace=True) #renombrar columna
print(iris.columns) #Verificación
```

## Modificar tablas

es posible hacer cambios de orden y estructura en las tablas de pandas.

```python
iris.sort_values(["sepal_width"], ascending=False)
```

Para reestructurar los datos usamos la función `pivot()`.

```python
pivot_iris=iris.pivot(columns='species',
                values=['sepal_width','sepal_length'])
pivot_iris.head()
```

## Concatenar y unir

Es posible unir varias tablas tanto vertical como horizontalmente.


En ambos casos, podemos usar `concat()`, y se juntarán las bases con bases a los nombres de columnas o los indices de las filas

```python
iris_setosa=iris[iris['species']=='setosa']
display(iris_setosa) #display:ejecuta el método dunder apropiado para obtener los datos apropiados para mostrar
```

```python
iris_virginica=iris[iris['species']=='virginica']
display(iris_virginica) #display:ejecuta el método dunder apropiado para obtener los datos apropiados para mostrar
```

Si se hace la concatenación sin más, se tomarán todas las columnas y se agregarán NaN

```python
iris_virginica_setosa=pd.concat([iris_setosa, iris_virginica])
display(iris_virginica_setosa)
```
